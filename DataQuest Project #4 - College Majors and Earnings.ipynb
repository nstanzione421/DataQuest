# College Major DeepDive

We'll explore how using the pandas plotting functionality along with the Jupyter notebook interface allows us to explore data quickly using visualizations.

We'll be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by American Community Survey, which conducts surveys and aggregates the data. FiveThirtyEight cleaned the dataset and released it on their Github repo.

Key questions we will look to address in this project:

* Do students in more popular majors make more money? Using scatter plots
* How many majors are predominantly male? Predominantly female? Using histograms
* Which category of majors have the most students? Using bar plots

We'll explore how to do these and more while primarily working in pandas. Before we start creating data visualizations, let's import the libraries we need and remove rows containing null values.

```python
import pandas as pd
import matplotlib as plt
%matplotlib inline
```

```python
file = 'recent-grads.csv'
recent_grads = pd.read_csv('/Users/nstanzione/Documents/EDU/DataQuest/Data/' + file)
```

```python
index = recent_grads.index
raw_data_count = len(index)
print(raw_data_count)
```

```python
recent_grads = recent_grads.dropna()
new_index = recent_grads.index
cleaned_data_count = len(new_index)
print(cleaned_data_count)
```

Looks like we had one row with null values. That has been removed from our dataframe and now we are good to go with the analysis. Let's start by exploring several different scatterplots and seeing if any relationships exist between some of the data fields.

```python
ax = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
ax.set_title('Sample Size v. Median Earnings')
```

```python
ax = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
ax.set_title('Sample Size v. Unemployment Rate')
```

```python
ax = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
ax.set_title('Employed Full Time v. Median Earnings')
```

```python
ax = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
ax.set_title('Pct Women v. Unemployment Rate')
```

```python
ax = recent_grads.plot(x='Men', y='Median', kind='scatter')
ax.set_title('Number of Men v. Median Earnings')
```

```python
ax = recent_grads.plot(x='Women', y='Median', kind='scatter')
ax.set_title('Number of Women v. Median Earnings')
```

```python
ax = recent_grads.plot(x='ShareWomen', y='Median', kind='scatter')
ax.set_title('Pct Women v. Median Earnings')
```

The charts above tell an interesting, and maybe expected, story. There look to be some weak relationship noting that less popular majors lead to higher earnings. The most recent chart shows that there appears to be a fairly strong relationship with majors with higher percentage share of women earn less.

Some of the initial charts were hard to review for the stroy such as using "Number of Men" and Number of Women." Those charts hid the relationship of women influence on earnings for majors. In addition, the full-time and unemployment charts showed no definitive relationships.

Next, lets look the distribution of some of the data fields using histograms.

```python
recent_grads['Sample_size'].hist(bins=20, range=(0,5000),grid=False)
ax.set_title('Sample Size')
```

```python
recent_grads['Median'].hist(bins=15, range=(0,150000),grid=False)
ax.set_title('Median Earnings')
```

```python
recent_grads['Employed'].hist(bins=10, range=(0,50000),grid=False)
ax.set_title('Number Employed')
```

```python
recent_grads['Full_time'].hist(bins=10, range=(0,50000),grid=False)
ax.set_title('Number Full-Time Employees')
```

```python
recent_grads['ShareWomen'].hist(bins=10, range=(0,1),grid=False)
ax.set_title('ShareWomen')
```

## Task 1: Update Columns
Let's first start by updating the column names to snakecase.

```python
autos.columns
```

```python
cols = ['dateCrawled', 'name', 'seller', 'offerType', 'price', 'abtest',
       'vehicleType', 'yearOfRegistration', 'gearbox', 'powerPS', 'model',
       'odometer', 'monthOfRegistration', 'fuelType', 'brand',
       'notRepairedDamage', 'dateCreated', 'nrOfPictures', 'postalCode',
       'lastSeen']

cols_clean = cols
cols_clean[7] = 'registration_year'
cols_clean[12] = 'registration_month'
cols_clean[15] = 'unrepaired_damage'
cols_clean[16] = 'ad_created'
cols_clean[17] = 'num_pictures'

print(cols_clean)
```

In the code above we idenitifed those names we wanted to change manually due to the fact they had multiple upper case words. Next, we will look to create two functions that we can use to update the remaining columns.

```python
def first(string):
    for i in range(0,len(string)):
        if (string[i].istitle()):
            return i

    return 0
    
def replace_cap(string):
    n = first(string)
    string_new = string[:n] + '_' + string[n].lower()+string[n+1:].lower()
    return string_new

test = 'testRun'
print(replace_cap(test))
    
```

The test above shows that our two new functions will be able to help us update the remaining column names. The first function has a string as an input, and then identifies the character in the string that has the first uppercase letter, and returns its position. The second function has a string as an input, that string is then fed through the first function, and the position of the upper case character is then used to adjust the string accordingly to be snakecase. We will loop through the remaining "bad" field names below and update the columns accordingly.

```python
bad_labels = [0,3,6,9,13,18,19]

for n in bad_labels:
    cols_clean[n] = replace_cap(cols_clean[n])

print(cols_clean)
```

```python
autos.columns = cols_clean
autos.head()
```

As the new dataframe headings show, we were able to successfully update the field names from camelcase to snakecase using the two functions described above after manually updating a few specific fields.

## Task 2: Updating price and odometer values to numbers

### Task 5: Brand Review
Let's dig a bit further into the brands most commonly sold on eBay based on our dataset.

```python
autos['brand'].value_counts()
```

The most popular brands are all European, further enforcing that the dat we are reviewing is foreign-based. Volkswagen is the top brand followed by Opel, and then the 3 popular luxury brands (BMW, Mercedes, Audi). Lada is the least common brand. 

Looking at the above list of brands, we will explore the top-10 brands as they account for ~80% of our data.

```python
autos['brand'].value_counts().index[:10]
```

```python
top10_brands = autos['brand'].value_counts().index[:10]
brand_price_avg = {}

for b in top10_brands:
    temp_df = autos[autos['brand']==b]
    mean_price = temp_df['price_usd'].mean()
    brand_price_avg[b] = mean_price

print(brand_price_avg)
```

As expected, our three luxury brands BMW, Mercedes, and Audi have the highest average sale price. All three have an average between 8K-10K USD (with Audi being thr most expensive. The Renault has the lowest average price at ~2,400 USD. Our most common brand, Volkswagen, is in the middle of the pack at ~5,300 USD.

```python
brand_odometer_avg = {}

for b in top10_brands:
    temp_df = autos[autos['brand']==b]
    mean_odometer = temp_df['odometer_km'].mean()
    brand_odometer_avg[b] = mean_odometer

print(brand_odometer_avg)
```

```python
avg_price = pd.Series(brand_price_avg)
avg_odometer = pd.Series(brand_odometer_avg)

top_brands = pd.DataFrame(avg_price,columns=['avg_price_usd'])
top_brands['avg_odometer_km'] = avg_odometer

pd.options.display.float_format = '{:,.0f}'.format

top_brands
```

There is no clear distinction of the mileage acorss the top-10 brands. If we rememebr from our earlier analysis, the odometer data might be a bit misleading with a potential maximum input value of 150,000. There is not logical correlation present between average mileage and average price for the brands. The brand itself is much more determinate of the price.
